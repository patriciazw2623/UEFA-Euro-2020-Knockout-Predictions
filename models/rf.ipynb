```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier
```

```python
TRAIN_FEATURE_URL = "https://raw.githubusercontent.com/weeeeiz/sc/main/21trainfeature.csv"
train_feature_path = tf.keras.utils.get_file("2021trainfeaturesssssssss.csv", TRAIN_FEATURE_URL)

TRAIN_LABEL_URL = "https://raw.githubusercontent.com/weeeeiz/sc/main/21trainlab.csv"
train_label_path = tf.keras.utils.get_file("2021---trainlabelss.csv", TRAIN_LABEL_URL)

TRAIN_ALL_URL = "https://raw.githubusercontent.com/weeeeiz/sc/main/21all.csv"
train_all_path = tf.keras.utils.get_file("2021alls.csv", TRAIN_ALL_URL)

TR_ALL_LABEL_URL = "https://raw.githubusercontent.com/weeeeiz/sc/main/21alllab.csv"
tr_all_label_path = tf.keras.utils.get_file("2021labs.csv", TR_ALL_LABEL_URL)

TEST_FEATURE_URL = "https://raw.githubusercontent.com/weeeeiz/sc/main/21testfeat.csv"
test_feature_path = tf.keras.utils.get_file("21_testfssss.csv", TEST_FEATURE_URL)

TEST_LABEL_URL = "https://raw.githubusercontent.com/weeeeiz/sc/main/21testlab.csv"
test_label_path = tf.keras.utils.get_file("21testlss.csv", TEST_LABEL_URL)

train_feature = pd.read_csv(train_all_path, header = None)
test_feature = pd.read_csv(test_feature_path, header = None)
train_label = pd.read_csv(tr_all_label_path, header = None)
test_label = pd.read_csv(test_label_path, header = None)
```

```python
model = RandomForestClassifier(n_estimators=25, max_depth = None, min_samples_split=2, random_state=1)
model.fit(train_feature, train_label)
```

```python
pred = model.predict(test_feature)
pred
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(test_label, pred))
```
